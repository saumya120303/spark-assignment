```python
from pyspark.sql import SparkSession
import glob

def create_spark_dataframe():
    spark = SparkSession.builder \
        .appName("COVID-19 Analysis") \
        .getOrCreate()

    # Read data from CSV into Spark DataFrame
    df_list = []
    csv_files = glob.glob('csvs/*_covid_data.csv')
    if len(csv_files) == 0:
        print("No CSV files found.")
        return None
    
    for file_name in csv_files:
        df = spark.read.csv(file_name, header=True)
        df_list.append(df)

    # Merge DataFrames into a single DataFrame
    if len(df_list) == 0:
        print("No data read from CSV files.")
        return None
    
    merged_df = df_list[0]
    for df in df_list[1:]:
        merged_df = merged_df.union(df)

    return merged_df

df = create_spark_dataframe()
if df is not None:
    df.show()
else:
    print("Failed to create Spark DataFrame.")
```
